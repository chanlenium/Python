```python
# 소상공인 경기동향(BSI) 지수 (소상공인, 부문별 실적 및 전망)
url = f'https://kosis.kr/openapi/Param/statisticsParameterData.do?method=getList&apiKey={apiKey}=&itmId=s0+s1+&objL1=ALL&objL2=&objL3=&objL4=&objL5=&objL6=&objL7=&objL8=&format=json&jsonVD=Y&prdSe=M&startPrdDe=201801&endPrdDe=202506&orgId=142&tblId=DT_S0001N_001'
r = requests.get(url, verify=False)
jo = r.json()
df = pd.DataFrame(jo)
temp_df = df[['TBL_NM', 'ITM_NM', 'PRD_DE', 'C1_NM', 'DT', 'UNIT_NM']].copy()

# 추출 데이터가 월단위 데이터이므로, 이를 Quater 단위로 변환
# 월 → 분기(Q1~Q4)로 변환하는 함수 : 3,6,9,12월 자료만 Quater로 변환하고 나머지는 Unknown 처리
def month_to_quarter(prd):
    year = prd[:4]
    month = prd[-2:]
    if month in ['03']:
        quarter = 'Q1'
    elif month in ['06']:
        quarter = 'Q2'
    elif month in ['09']:
        quarter = 'Q3'
    elif month in ['12']:
        quarter = 'Q4'
    else:
        quarter = 'Unknown'
    return f"{year}{quarter}"

# API Data를 DataFrame형태로 저장할 딕셔너리
split_dfs = {}

# 조건 기준들
filter_1 = ['체감', '전망']
filter_2 = ['경기전반', '매출', '영업이익', '자금사정', '원재료조달사정', '재고', '고용', '비용상황', '구매고객수']

# 조건 조합 반복
for s in filter_1:
    for p in filter_2:
        # 앞에 이름 8개만 잘라서 Data를 담고있는 'DT'컬럼의 이름을 대체함
        # (예) '소상공인 부문별_비용상황_체감' => '소상공인_부문별'
        newColName = f"{temp_df['TBL_NM'].unique()[0][:8]}_{p}_{s}"        
        filtered = temp_df[(temp_df['ITM_NM'] == s) & (temp_df['C1_NM'] == p)]
        split_dfs[newColName] = filtered.copy()
        split_dfs[newColName].rename(columns={'DT': newColName}, inplace=True)
        split_dfs[newColName]['PRD_DE'] = split_dfs[newColName]['PRD_DE'].apply(month_to_quarter)
        # Unknown이 들어가있는 행을 삭제하여 Quater 데이터만 남김
        split_dfs[newColName] = split_dfs[newColName][~split_dfs[newColName]['PRD_DE'].str.contains('Unknown')]
        
# split_dfs['소상공인 부문별_비용상황_체감']

# 데이터마다 시계열이 다르므로, 공통 시계열(2019Q1 ~ 2022Q4) DataFrame을 만들어서 추출된 데이터를 붙임(Left Join) 
# 연도 및 분기 리스트
years = range(2018, 2023)  # 2018 ~ 2022
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
# PRD_DE 값 생성
prd_de_list = [f"{year}{q}" for year in years for q in quarters]
# DataFrame 생성 (열 이름이 'PRD_DE')
df = pd.DataFrame({'PRD_DE': prd_de_list})

# DataFrame Join
df_merged = df.copy()
for name, sub in split_dfs.items():
    df_merged = df_merged.merge(sub.iloc[:, [2, 4]], on='PRD_DE', how='left', suffixes=('', '_dup'))  # 필요에 따라 'outer' 등으로 변경
df_merged
#df_merged.to_excel("output.xlsx")
```
